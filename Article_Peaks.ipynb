```python
# import packages
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import scienceplots
import nltk
nltk.download('punkt')
```

```python
# Read ukraine data
u_data = pd.read_csv('drive/MyDrive/Master Thesis/ukraine_complete.csv')
```

```python
# Read syria data
s_data = pd.read_csv('drive/MyDrive/Master Thesis/syria_complete.csv')
```

```python
# Drop column Unnamed 
u_data = u_data.iloc[: , 1:]
s_data = s_data.iloc[: , 1:]
```

```python
s_data.head()
```

```python
u_data.head()
```

## Ukraine


```python
# Convert the Publication Date column to a datetime format
u_data['Date'] = pd.to_datetime(u_data['Date'])

# Group the data by Publication Date and count the number of articles in each group
grouped = u_data.groupby(u_data['Date'].dt.date).count()

# Plot the number of articles against the Publication Date
plt.figure(figsize=(30,10))
plt.plot(grouped.index, grouped['Date'])

plt.xlabel('Publication Date')
plt.ylabel('Number of Articles')
plt.title('Ukraine Number of Articles vs. Publication Date', fontsize = 20)
plt.show()
```
